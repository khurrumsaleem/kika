# KIKA Materials Module - Complete Tutorial

This notebook demonstrates all the functionality of the KIKA materials module, which provides a general-purpose framework for defining, manipulating, and exporting material compositions for various Monte Carlo codes including MCNP and Serpent.

```python
from kika.materials import Material, MaterialCollection, Nuclide, NuclideAccessor
```

## 1. Creating Materials

Materials can be created in several ways. The most basic is to create an empty material and add nuclides to it.

### 1.1 Basic Material Creation

```python
# Create a material with ID=1
water = Material(id=1)

# Add hydrogen and oxygen (natural elements)
water.add_nuclide('H', 2.0, 'ao')  # 2 atoms of H
water.add_nuclide('O', 1.0, 'ao')  # 1 atom of O

# Set the name and density
water.set_name('Water')
water.set_density(1.0, 'g/cc')

print(water)
```

### 1.2 Adding Nuclides by ZAID

You can also add nuclides by their ZAID (ZZAAA format where ZZ is atomic number and AAA is mass number).

```python
# Create enriched uranium
uranium = Material(id=2)
uranium.set_name('Enriched Uranium')

# Add by ZAID (92235 = U-235, 92238 = U-238)
uranium.add_nuclide(92235, 0.05, 'ao')  # 5% U-235
uranium.add_nuclide(92238, 0.95, 'ao')  # 95% U-238

uranium.set_density(19.1, 'g/cc')

print(uranium)
```

### 1.3 Adding Natural Elements with add_element()

The `add_element()` method is a convenience function for adding natural elements.

```python
# Create stainless steel (simplified)
steel = Material(id=3)
steel.set_name('Stainless Steel 304')

# Add natural elements
steel.add_element('Fe', 0.70, 'wo')  # 70% iron by weight
steel.add_element('Cr', 0.18, 'wo')  # 18% chromium
steel.add_element('Ni', 0.10, 'wo')  # 10% nickel
steel.add_element('C', 0.02, 'wo')   # 2% carbon

steel.set_density(8.0, 'g/cc')

print(steel)
```

### 1.4 Setting Temperature

```python
# Set temperature in Kelvin
water.set_temperature(300.0)  # 300 K (~27°C)
print(f"Water temperature: {water.temperature} K")
```

## 2. Accessing Material Properties

### 2.1 Nuclide Access by Symbol or ZAID

```python
# Access by symbol
h_nuclide = water.nuclide['H']
print(f"Hydrogen by symbol: ZAID={h_nuclide.zaid}, fraction={h_nuclide.fraction}")

# Access by ZAID
o_nuclide = water.nuclide[8000]  # 8000 = natural oxygen
print(f"Oxygen by ZAID: symbol={o_nuclide.symbol}, fraction={o_nuclide.fraction}")

# Check if nuclide exists
print(f"\nHas 'Fe'? {'Fe' in water.nuclide}")
print(f"Has 'H'? {'H' in water.nuclide}")
```

### 2.2 Density with Units

```python
# density property returns just the float
print(f"Density (float): {water.density}")

# density_with_units property returns a formatted string
print(f"Density (with units): {water.density_with_units}")

# Convert density to different units
print(f"Density in kg/m3: {water.density_in('kg/m3')}")
```

### 2.3 Fraction Type

```python
print(f"Water fraction type: {water.fraction_type}")
print(f"Water is_atomic: {water.is_atomic}")
print(f"Water is_weight: {water.is_weight}")

print(f"\nSteel fraction type: {steel.fraction_type}")
print(f"Steel is_atomic: {steel.is_atomic}")
print(f"Steel is_weight: {steel.is_weight}")
```

## 3. Modifying Materials

### 3.1 Setter Methods

```python
# Create a test material
test_mat = Material(id=100)

# set_name()
test_mat.set_name('Test Material')
print(f"Name: {test_mat.name}")

# set_id()
test_mat.set_id(200)
print(f"ID: {test_mat.id}")

# set_library()
test_mat.set_library('nlib', '80c')
print(f"Libraries: {test_mat.libs}")

# remove_library()
test_mat.remove_library('nlib')
print(f"Libraries after removal: {test_mat.libs}")
```

### 3.2 Nuclide Modification Methods

```python
# Create a material
mod_mat = Material(id=10)
mod_mat.add_nuclide('H', 1.0, 'ao')
mod_mat.add_nuclide('O', 1.0, 'ao')
mod_mat.add_nuclide('Fe', 1.0, 'ao')

print("Initial nuclides:")
for sym, nuc in mod_mat.nuclide.items():
    print(f"  {sym}: {nuc.fraction}")

# update_nuclide_fraction()
mod_mat.update_nuclide_fraction('H', 2.0)
print(f"\nAfter update_nuclide_fraction('H', 2.0):")
print(f"  H: {mod_mat.nuclide['H'].fraction}")

# remove_nuclide()
mod_mat.remove_nuclide('Fe')
print(f"\nAfter remove_nuclide('Fe'):")
for sym, nuc in mod_mat.nuclide.items():
    print(f"  {sym}: {nuc.fraction}")

# clear_nuclides()
mod_mat.clear_nuclides()
print(f"\nAfter clear_nuclides():")
print(f"  Number of nuclides: {len(mod_mat.nuclide)}")
```

### 3.3 copy() - Create Modified Versions

```python
# Create a copy of water with a new ID
heavy_water = water.copy(new_id=11)
heavy_water.set_name('Heavy Water')

# Modify the copy (replace H with D)
heavy_water.remove_nuclide('H')
heavy_water.add_nuclide(1002, 2.0, 'ao')  # Deuterium
heavy_water.set_density(1.1, 'g/cc')

print(f"Original water ID: {water.id}, nuclides: {list(water.nuclide.keys())}")
print(f"Heavy water ID: {heavy_water.id}, nuclides: {list(heavy_water.nuclide.keys())}")
```

## 4. Fraction Conversions

### 4.1 to_atomic_fraction() / to_weight_fraction()

```python
# Create water with atomic fractions
water_test = Material(id=20)
water_test.add_nuclide('H', 2.0, 'ao')
water_test.add_nuclide('O', 1.0, 'ao')

print("Original (atomic fractions):")
print(f"  Type: {water_test.fraction_type}")
for sym, nuc in water_test.nuclide.items():
    print(f"  {sym}: {nuc.fraction:.6f}")

# Convert to weight fractions
water_test.to_weight_fraction()
print("\nAfter to_weight_fraction():")
print(f"  Type: {water_test.fraction_type}")
for sym, nuc in water_test.nuclide.items():
    print(f"  {sym}: {nuc.fraction:.6f}")

# Convert back to atomic fractions
water_test.to_atomic_fraction()
print("\nAfter to_atomic_fraction():")
print(f"  Type: {water_test.fraction_type}")
for sym, nuc in water_test.nuclide.items():
    print(f"  {sym}: {nuc.fraction:.6f}")
```

### 4.2 normalize() - Scale Fractions to Sum=1

```python
# Create a material with arbitrary fractions
h2o = Material(id=21)
h2o.add_nuclide('H', 2.0, 'ao')  # Not normalized (sum = 3.0)
h2o.add_nuclide('O', 1.0, 'ao')

print("Before normalize():")
total = sum(nuc.fraction for nuc in h2o.nuclide.values())
print(f"  Total: {total}")
for sym, nuc in h2o.nuclide.items():
    print(f"  {sym}: {nuc.fraction:.6f}")

# Normalize to atomic fractions (sum = 1.0)
h2o.normalize('ao')

print("\nAfter normalize('ao'):")
total = sum(nuc.fraction for nuc in h2o.nuclide.values())
print(f"  Total: {total}")
for sym, nuc in h2o.nuclide.items():
    print(f"  {sym}: {nuc.fraction:.6f}")
```

### 4.3 to_integer_fractions() - Convert to Integer Ratios (NEW)

This method converts normalized fractions back to integer ratios. For example, H2O with fractions (H: 0.6667, O: 0.3333) becomes (H: 2, O: 1).

```python
# Create a normalized material
mat_norm = Material(id=22)
mat_norm.add_nuclide('H', 2.0/3.0, 'ao')  # Normalized H2O
mat_norm.add_nuclide('O', 1.0/3.0, 'ao')

print("Normalized fractions:")
for sym, nuc in mat_norm.nuclide.items():
    print(f"  {sym}: {nuc.fraction:.6f}")

# Convert to integer fractions
mat_norm.to_integer_fractions()

print("\nAfter to_integer_fractions():")
for sym, nuc in mat_norm.nuclide.items():
    print(f"  {sym}: {nuc.fraction:.1f}")
```

```python
# More complex example: sulfuric acid H2SO4
h2so4 = Material(id=23)
h2so4.add_nuclide('H', 2.0, 'ao')
h2so4.add_nuclide('S', 1.0, 'ao')
h2so4.add_nuclide('O', 4.0, 'ao')
h2so4.normalize('ao')

print("H2SO4 normalized:")
for sym, nuc in sorted(h2so4.nuclide.items()):
    print(f"  {sym}: {nuc.fraction:.6f}")

h2so4.to_integer_fractions()
print("\nH2SO4 integer fractions:")
for sym, nuc in sorted(h2so4.nuclide.items()):
    print(f"  {sym}: {nuc.fraction:.0f}")
```

## 5. Natural Element Expansion

### 5.1 expand_natural_elements() - All Elements

```python
# Create a material with natural elements
nat_mat = Material(id=30)
nat_mat.add_nuclide('H', 2.0, 'ao')
nat_mat.add_nuclide('O', 1.0, 'ao')

print("Before expansion:")
for sym, nuc in nat_mat.nuclide.items():
    print(f"  {sym} (ZAID {nuc.zaid}): {nuc.fraction:.6f}")

# Expand all natural elements
nat_mat.expand_natural_elements()

print("\nAfter expand_natural_elements():")
for sym, nuc in sorted(nat_mat.nuclide.items()):
    print(f"  {sym} (ZAID {nuc.zaid}): {nuc.fraction:.6e}")
```

### 5.2 expand_natural_elements() - Selective Expansion

```python
# Create steel with natural elements
steel2 = Material(id=31)
steel2.add_element('Fe', 0.70, 'ao')
steel2.add_element('Cr', 0.18, 'ao')
steel2.add_element('Ni', 0.10, 'ao')
steel2.add_element('C', 0.02, 'ao')

# Expand only Fe
steel2.expand_natural_elements('Fe')
print("After expanding only Fe:")
for sym, nuc in sorted(steel2.nuclide.items(), key=lambda x: x[1].zaid):
    print(f"  {sym}: {nuc.fraction:.6e}")
```

```python
# Expand specific elements with a list
steel3 = Material(id=32)
steel3.add_element('Fe', 0.70, 'ao')
steel3.add_element('Cr', 0.18, 'ao')
steel3.add_element('Ni', 0.10, 'ao')
steel3.add_element('C', 0.02, 'ao')

# Expand Fe and C only
steel3.expand_natural_elements(['Fe', 'C'])
print("After expanding Fe and C:")
for sym, nuc in sorted(steel3.nuclide.items(), key=lambda x: x[1].zaid):
    print(f"  {sym}: {nuc.fraction:.6e}")
```

## 6. Code Output Formats

### 6.1 to_mcnp() - MCNP Material Cards

```python
# Create a complete material for MCNP output
mcnp_mat = Material(id=100)
mcnp_mat.set_name('Test Material')
mcnp_mat.add_nuclide('U235', 0.05, 'ao', '80c')  # With library
mcnp_mat.add_nuclide('U238', 0.95, 'ao', '80c')
mcnp_mat.set_density(19.1, 'g/cc')

print("MCNP output:")
print(mcnp_mat.to_mcnp())
```

### 6.2 to_serpent() - Serpent Material Definition (NEW)

```python
# Same material in Serpent format
print("Serpent output:")
print(mcnp_mat.to_serpent())
```

```python
# Serpent output with temperature
hot_mat = Material(id=101)
hot_mat.set_name('Hot Water')
hot_mat.add_nuclide('H', 2.0, 'ao', '80c')
hot_mat.add_nuclide('O', 1.0, 'ao', '80c')
hot_mat.normalize()
hot_mat.set_density(0.95, 'g/cc')
hot_mat.set_temperature(600.0)  # 600 K

print("Serpent output with temperature:")
print(hot_mat.to_serpent())
```

### 6.3 Weight Fraction Output

```python
# Weight fraction material
wo_mat = Material(id=102)
wo_mat.add_nuclide('Fe', 0.70, 'wo')
wo_mat.add_nuclide('Cr', 0.18, 'wo')
wo_mat.add_nuclide('Ni', 0.12, 'wo')
wo_mat.set_density(8.0, 'g/cc')

print("MCNP output (weight fractions):")
print(wo_mat.to_mcnp())
print("\nSerpent output (weight fractions):")
print(wo_mat.to_serpent())
```

## 7. Material Collections

### 7.1 Creating a Collection

```python
# Create materials
mat1 = Material(id=1)
mat1.set_name('Water')
mat1.add_nuclide('H', 2.0, 'ao')
mat1.add_nuclide('O', 1.0, 'ao')
mat1.normalize()
mat1.set_density(1.0, 'g/cc')

mat2 = Material(id=2)
mat2.set_name('Uranium')
mat2.add_nuclide(92235, 0.05, 'ao')
mat2.add_nuclide(92238, 0.95, 'ao')
mat2.set_density(19.1, 'g/cc')

# Create collection and add materials
collection = MaterialCollection()
collection.add_material(mat1)
collection.add_material(mat2)

print(collection)
```

### 7.2 Accessing Materials in Collection

```python
# Access by ID
water = collection.by_id[1]
print(f"Material 1: {water.name}")

# Iterate over all materials
print("\nAll materials:")
for mat in collection:
    print(f"  ID {mat.id}: {mat.name}")
```

### 7.3 Collection Export

```python
# Export all materials as MCNP cards
print("=== MCNP Output ===")
print(collection.to_mcnp())
```

```python
# Export all materials as Serpent cards
print("=== Serpent Output ===")
print(collection.to_serpent())
```

## 8. Summary of New Features

The materials module has been restructured and enhanced with the following new features:

### Module Structure
- Moved from `kika.mcnp.material` to `kika.materials` (with backwards compatibility)
- Supports multi-code output (MCNP, Serpent)

### New Properties
- `density_with_units` - Returns density as a formatted string with units

### New Methods
- `to_serpent()` - Export material as Serpent format
- `to_integer_fractions()` - Convert normalized fractions to integer ratios
- `set_name()` - Set or change material name
- `set_id()` - Change material ID
- `set_library()` - Set material-level library defaults
- `remove_library()` - Remove material-level library setting
- `remove_nuclide()` - Remove a nuclide from the material
- `update_nuclide_fraction()` - Update fraction of existing nuclide
- `clear_nuclides()` - Remove all nuclides

### Collection Methods
- `to_serpent()` - Export all materials as Serpent format
