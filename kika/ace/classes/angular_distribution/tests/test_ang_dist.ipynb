```python
ace_file = '/share_snc/snc/JuanMonleon/plot_samples_leg/260560_40.02c'
ace1_file = '/home/MONLEON-JUAN/KIKA/kika/sampling/test_endf_pert/ace/293.6/26056/0001/260560_33_0001.02c'
ace2_file = '/home/MONLEON-JUAN/KIKA/kika/sampling/test_endf_pert/ace/293.6/26056/0002/260560_33_0002.02c'

ace = kika.read_ace(ace_file)
ace1 = kika.read_ace(ace1_file)
ace2 = kika.read_ace(ace2_file)
```

```python
# Compare ACE angular distributions using PlotBuilder
energy = 0.755
labels = ['ref', 'sample01', 'sample02']
colors = ['C0', 'C1', 'C2']

builder = PlotBuilder(style='light', figsize=(10, 6))

for ace_obj, label, color in zip([ace, ace1, ace2], labels, colors):
    data = ace_obj.angular_distributions.to_plot_data(
        mt=2, energy=energy, ace=ace_obj, label=label, color=color
    )
    builder.add_data(data)

fig = (builder
       .set_labels(
           title=f'Angular Distribution Comparison - MT=2 at {energy} MeV',
           x_label=r'Cosine ($\mu$)',
           y_label='Probability Density'
       )
       .set_legend(loc='best')
       .build())
```
